```python
import pandas as pd
from sklearn.decomposition import PCA, FastICA, TruncatedSVD
from sklearn.preprocessing import StandardScaler
import warnings
from sklearn.exceptions import ConvergenceWarning

# Suppress FastICA convergence warnings
warnings.filterwarnings("ignore", category=ConvergenceWarning)

# -----------------------------
# 1. Breast Cancer Dataset
# -----------------------------
breast_path = "/kaggle/input/breast-cancer-wisconsin-data/data.csv"
breast_df = pd.read_csv(breast_path)
breast_df_clean = breast_df.drop(columns=['id', 'Unnamed: 32'])
breast_df_clean['diagnosis'] = breast_df_clean['diagnosis'].map({'M':1, 'B':0})

X_breast = breast_df_clean.values
scaler = StandardScaler()
X_breast_scaled = scaler.fit_transform(X_breast)

# PCA
pca = PCA()
X_breast_pca = pca.fit_transform(X_breast_scaled)
explained_var = pca.explained_variance_ratio_
cum_var = explained_var.cumsum()
n_pca = (cum_var < 0.95).sum() + 1
print(f"Breast Cancer PCA components covering 95% variance: {n_pca}")

# ICA
ica = FastICA(n_components=n_pca, random_state=42, max_iter=5000, tol=1e-6)
X_breast_ica = ica.fit_transform(X_breast_scaled)

# SVD
svd = TruncatedSVD(n_components=n_pca, random_state=42)
X_breast_svd = svd.fit_transform(X_breast_scaled)

print("Shape of PCA features:", X_breast_pca.shape)
print("Shape of ICA features:", X_breast_ica.shape)
print("Shape of SVD features:", X_breast_svd.shape)

# -----------------------------
# 2. Boston Housing Dataset
# -----------------------------
boston_path = "/root/.cache/kagglehub/datasets/simpleparadox/bostonhousingdataset/versions/1/boston-housing-dataset.csv"
boston_df = pd.read_csv(boston_path)
boston_df_clean = boston_df.drop(columns=['Unnamed: 0'])
X_boston = boston_df_clean.values
X_boston_scaled = scaler.fit_transform(X_boston)

# PCA
pca = PCA()
X_boston_pca = pca.fit_transform(X_boston_scaled)
explained_var = pca.explained_variance_ratio_
cum_var = explained_var.cumsum()
n_pca = (cum_var < 0.95).sum() + 1
print(f"Boston Housing PCA components covering 95% variance: {n_pca}")

# ICA
ica = FastICA(n_components=n_pca, random_state=42, max_iter=5000, tol=1e-6)
X_boston_ica = ica.fit_transform(X_boston_scaled)

# SVD
svd = TruncatedSVD(n_components=n_pca, random_state=42)
X_boston_svd = svd.fit_transform(X_boston_scaled)

print("Shape of PCA features:", X_boston_pca.shape)
print("Shape of ICA features:", X_boston_ica.shape)
print("Shape of SVD features:", X_boston_svd.shape)

# -----------------------------
# 3. Iris Dataset
# -----------------------------
iris_path = "/kaggle/input/iris/Iris.csv"
iris_df = pd.read_csv(iris_path)
iris_df_clean = iris_df.drop(columns=['Id', 'Species'])
X_iris = iris_df_clean.values
X_iris_scaled = scaler.fit_transform(X_iris)

# PCA
pca = PCA()
X_iris_pca = pca.fit_transform(X_iris_scaled)
explained_var = pca.explained_variance_ratio_
cum_var = explained_var.cumsum()
n_pca = (cum_var < 0.95).sum() + 1
print(f"Iris PCA components covering 95% variance: {n_pca}")

# ICA
ica = FastICA(n_components=n_pca, random_state=42, max_iter=5000, tol=1e-6)
X_iris_ica = ica.fit_transform(X_iris_scaled)

# SVD
svd = TruncatedSVD(n_components=n_pca, random_state=42)
X_iris_svd = svd.fit_transform(X_iris_scaled)

print("Shape of PCA features:", X_iris_pca.shape)
print("Shape of ICA features:", X_iris_ica.shape)
print("Shape of SVD features:", X_iris_svd.shape)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import time

# Load and preprocess Breast Cancer dataset
breast_path = "/kaggle/input/breast-cancer-wisconsin-data/data.csv"
breast_df = pd.read_csv(breast_path)
breast_df_clean = breast_df.drop(columns=['id', 'Unnamed: 32'])
breast_df_clean['diagnosis'] = breast_df_clean['diagnosis'].map({'M':1, 'B':0})

X_original = breast_df_clean.drop(columns=['diagnosis']).values
y = breast_df_clean['diagnosis'].values

# Split data
X_train_orig, X_test_orig, y_train, y_test = train_test_split(X_original, y, test_size=0.2, random_state=42)

# Standardize
scaler = StandardScaler()
X_train_orig = scaler.fit_transform(X_train_orig)
X_test_orig = scaler.transform(X_test_orig)

# --- Models to train ---
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=42)
}

# Function to train and evaluate
def train_evaluate(X_train, X_test, y_train, y_test, model_dict):
    results = {}
    for name, model in model_dict.items():
        start_time = time.time()
        model.fit(X_train, y_train)
        train_time = time.time() - start_time
        y_pred = model.predict(X_test)
        results[name] = {
            "accuracy": accuracy_score(y_test, y_pred),
            "precision": precision_score(y_test, y_pred),
            "recall": recall_score(y_test, y_pred),
            "f1": f1_score(y_test, y_pred),
            "train_time": train_time
        }
    return results

# 1. Original features
results_original = train_evaluate(X_train_orig, X_test_orig, y_train, y_test, models)

# 2. Extracted features (use previously computed PCA features, for example)
X_pca = X_breast_pca  # from previous PCA step
X_train_pca, X_test_pca, _, _ = train_test_split(X_pca, y, test_size=0.2, random_state=42)
results_pca = train_evaluate(X_train_pca, X_test_pca, y_train, y_test, models)

# 3. Selected features (SelectKBest)
X_skb = X_breast_skb  # from previous step
X_train_skb, X_test_skb, _, _ = train_test_split(X_skb, y, test_size=0.2, random_state=42)
results_skb = train_evaluate(X_train_skb, X_test_skb, y_train, y_test, models)

# Display results
print("=== Breast Cancer Model Performance ===")
print("Original Features:\n", results_original)
print("\nPCA Features:\n", results_pca)
print("\nSelected Features (SelectKBest/RFE):\n", results_skb)
```

```python
from sklearn.linear_model import SGDRegressor
import numpy as np
import matplotlib.pyplot as plt

# Custom fit loop to track convergence and print results
def sgd_track_verbose(X_train, y_train, max_iter=1000, tol=1e-4, print_every=100):
    sgd = SGDRegressor(max_iter=1, learning_rate='invscaling', eta0=0.01,
                       tol=None, random_state=42, warm_start=True)
    losses = []
    for i in range(max_iter):
        sgd.fit(X_train, y_train)
        y_pred = sgd.predict(X_train)
        loss = np.mean((y_train - y_pred)**2)
        losses.append(loss)
        # Print progress every 'print_every' iterations
        if (i+1) % print_every == 0 or i == 0:
            print(f"Iteration {i+1}: Training MSE = {loss:.4f}")
        # Stop early if loss change is below tolerance
        if i > 0 and abs(losses[-2] - loss) < tol:
            print(f"Converged at iteration {i+1} with MSE = {loss:.4f}")
            break
    return losses

# Track SGD for original and PCA features
loss_orig_track = sgd_track_verbose(X_train, y_train, max_iter=1000)
loss_pca_track = sgd_track_verbose(X_train_pca, y_train, max_iter=1000)

# Print final MSE for summary
print("\nFinal Training MSE:")
print(f"Original features: {loss_orig_track[-1]:.4f}")
print(f"PCA features:      {loss_pca_track[-1]:.4f}")

# Plot convergence
plt.figure(figsize=(8,5))
plt.plot(loss_orig_track, label='Original Features')
plt.plot(loss_pca_track, label='PCA Features')
plt.xlabel("Iteration")
plt.ylabel("Training MSE")
plt.title("SGDRegressor Convergence: Original vs PCA Features")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import time

# Load Iris dataset
iris_path = "/kaggle/input/iris/Iris.csv"
iris_df = pd.read_csv(iris_path)
iris_df_clean = iris_df.drop(columns=['Id', 'Species'])
X_original = iris_df_clean.values

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_original)

# PCA reduction to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Function to run KMeans and track results
def run_kmeans(X, n_clusters=3, random_state=42):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
    start_time = time.time()
    kmeans.fit(X)
    elapsed_time = time.time() - start_time
    inertia = kmeans.inertia_
    silhouette = silhouette_score(X, kmeans.labels_)
    n_iter = kmeans.n_iter_
    return inertia, silhouette, n_iter, elapsed_time

# Run KMeans on original data
inertia_orig, silhouette_orig, n_iter_orig, time_orig = run_kmeans(X_scaled)

# Run KMeans on PCA-reduced data
inertia_pca, silhouette_pca, n_iter_pca, time_pca = run_kmeans(X_pca)

# Print results
print("=== KMeans Clustering Performance ===")
print("Original Features:")
print(f"Inertia: {inertia_orig:.4f}, Silhouette Score: {silhouette_orig:.4f}, Iterations: {n_iter_orig}, Time: {time_orig:.4f}s")

print("\nPCA Features (2 components):")
print(f"Inertia: {inertia_pca:.4f}, Silhouette Score: {silhouette_pca:.4f}, Iterations: {n_iter_pca}, Time: {time_pca:.4f}s")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import time

# -----------------------------
# Load and preprocess Breast Cancer dataset
# -----------------------------
breast_path = "/kaggle/input/breast-cancer-wisconsin-data/data.csv"
breast_df = pd.read_csv(breast_path)
breast_df_clean = breast_df.drop(columns=['id', 'Unnamed: 32'])
breast_df_clean['diagnosis'] = breast_df_clean['diagnosis'].map({'M':1, 'B':0})

X_original = breast_df_clean.drop(columns=['diagnosis']).values
y = breast_df_clean['diagnosis'].values

# Use previously computed PCA features and SelectKBest features
X_pca = X_breast_pca  # PCA-reduced features
X_selected = X_breast_skb  # SelectKBest features

# Standardize original features
scaler = StandardScaler()
X_original_scaled = scaler.fit_transform(X_original)

# Split data for all versions
X_train_orig, X_test_orig, y_train, y_test = train_test_split(X_original_scaled, y, test_size=0.2, random_state=42)
X_train_pca, X_test_pca, _, _ = train_test_split(X_pca, y, test_size=0.2, random_state=42)
X_train_sel, X_test_sel, _, _ = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# -----------------------------
# 1. KNN classifier (distance-based)
# -----------------------------
def evaluate_knn(X_train, X_test, y_train, y_test, n_neighbors=5):
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    start_time = time.time()
    knn.fit(X_train, y_train)
    fit_time = time.time() - start_time

    start_time = time.time()
    y_pred = knn.predict(X_test)
    predict_time = time.time() - start_time
    acc = accuracy_score(y_test, y_pred)
    return acc, fit_time, predict_time

# KNN on original and PCA data
knn_orig = evaluate_knn(X_train_orig, X_test_orig, y_train, y_test)
knn_pca = evaluate_knn(X_train_pca, X_test_pca, y_train, y_test)

print("=== KNN Classifier Performance ===")
print(f"Original Features: Accuracy={knn_orig[0]:.4f}, FitTime={knn_orig[1]:.4f}s, PredictTime={knn_orig[2]:.4f}s")
print(f"PCA Features:      Accuracy={knn_pca[0]:.4f}, FitTime={knn_pca[1]:.4f}s, PredictTime={knn_pca[2]:.4f}s")

# -----------------------------
# 2. Random Forest classifier (tree-based)
# -----------------------------
def evaluate_rf(X_train, X_test, y_train, y_test):
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    start_time = time.time()
    rf.fit(X_train, y_train)
    fit_time = time.time() - start_time

    y_pred = rf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    return acc, fit_time

rf_orig = evaluate_rf(X_train_orig, X_test_orig, y_train, y_test)
rf_pca = evaluate_rf(X_train_pca, X_test_pca, y_train, y_test)
rf_sel = evaluate_rf(X_train_sel, X_test_sel, y_train, y_test)

print("\n=== Random Forest Classifier Performance ===")
print(f"Original Features: Accuracy={rf_orig[0]:.4f}, FitTime={rf_orig[1]:.4f}s")
print(f"PCA Features:      Accuracy={rf_pca[0]:.4f}, FitTime={rf_pca[1]:.4f}s")
print(f"Selected Features: Accuracy={rf_sel[0]:.4f}, FitTime={rf_sel[1]:.4f}s")
```
